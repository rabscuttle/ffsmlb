```python
# Import
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from time import sleep
```

```python
# Initialize Browser

# Firefox
options = Options()
options.binary_location = r'C:\Program Files\Mozilla Firefox\firefox.exe'
options.headless = True
firefox_binary = FirefoxBinary()
browser = webdriver.Firefox(executable_path=r'C:\bin\geckodriver.exe', options=options, firefox_binary=firefox_binary)

# Chrome
#executable_path = {'executable_path': 'c:/bin/chromedriver.exe'}
#browser = Browser('chrome', **executable_path, headless=True)
```

```python
# Read csv
player_list = pd.read_csv('input files/playerinfo.csv', encoding = "ISO-8859-1")
name = player_list['name']
urls = player_list['url']
owner = player_list['owner']
current_year = '2024'
```

```python
# Declare Variables
player_table = []
```

```python
loop_counter = 0
for url in urls:
    
    # Variables
    # Pitching
    total_w = 0
    total_sv = 0
    total_hld = 0
    total_ip = 0
    total_bf = 0
    total_ha = 0
    total_er = 0
    total_so = 0
    total_war = 0
    total_bba = 0
    total_ibba = 0
    total_hbpa = 0
    total_hra = 0
    full_ip = 0
    partial_ip = 0
    fb_velo = 0
    total_war = 0
    total_gb = 0
    total_fb = 0
    total_ld = 0
    siera = 0
    #Hitting
    total_h_ab = 0
    total_h_pa = 0
    total_singles = 0
    total_doubles = 0
    total_triples = 0
    total_hrs = 0
    total_h_r = 0
    total_rbi = 0
    total_h_bb = 0
    total_h_ibb = 0
    total_h_so = 0
    total_hbp = 0
    total_sf = 0
    total_sb = 0
    total_events = 0
    total_barrels = 0
    total_hardhit = 0
    hardhit_percent = 0
    barrel_percent = 0
    total_wrc = 0
    ev = 0
    la = 0
    position = ''
    team = ''

    #Constants
    lgHR_FB_percent = .114 # https://www.fangraphs.com/leaders.aspx?pos=all&stats=pit&lg=all&qual=0&type=2&season=2022&month=0&season1=2022&ind=0&team=0,ss&rost=0&age=&filter=&players=0&startdate=&enddate=
    cFIP = 3.117 # This item and below found at: https://www.fangraphs.com/guts.aspx?type=cn
    wbb = .689
    whbp = .721
    w1b = .884
    w2b = 1.26
    w3b = 1.599
    whr = 2.067
    league_woba = .310
    woba_scale = 1.257
    r_pa = .115
    
    # Player Link and DP Year
    fg_link = url
    
    # Get Soup
    try:
        # Firefox
        browser.get(fg_link)
        html = browser.page_source
        sleep(10)
        
        # Chrome
        #browser.visit(fg_link)
        #html = browser.html
        
        soup = bs(html, 'html.parser')
        # Get Player Info
        fg_name = soup.select('#content > div > div.player-pages-layout_player-pages-layout__content__U2dUD > div:nth-child(1) > div.header_header__wrapper__4W49n > div > div > div.header_player-info__top__UwQPj > div.header_player-name__kZWhq > h1')[0].text
        team_level = soup.select('#content > div > div.player-pages-layout_player-pages-layout__content__U2dUD > div:nth-child(1) > div.header_header__wrapper__4W49n > div > div > div.header_player-info__top__UwQPj > div.header_player-name__kZWhq')[0].text
        position = soup.select('#content > div > div.player-pages-layout_player-pages-layout__content__U2dUD > div:nth-child(1) > div.header_header__wrapper__4W49n > div > div > div.header_item__6AFbi.header_item__position__tADVT.header_is-desktop__PVkUu')[0].text
    
        if position == "PH/PR":
            position = "DH"
            
        #team = team_level.split('(')[0]
        try:
            level = team_level.split('(')[1].replace(')', '')
        except:
            level = ""
        player_dob = (soup.select('#content > div > div.player-pages-layout_player-pages-layout__content__U2dUD > div:nth-child(1) > div.header_header__wrapper__4W49n > div > div > div.header_player-info__data__CtgH_ > div.header_player-info__data__bottom__iSpVl > div > div:nth-child(1)')[0].text).split(' ')[1]
    except:
        fg_name = ""
        player_dob = ""
    
    #################################
    ####### P I T C H E R S #########
    #################################
    if position == 'P':
    #if 'P' in position:
        
#         # Prospects Report - HAVE TO FIGURE OUT THE LENGTH (IN CELLS) OF THE BOX BECAUSE IT IS VARIABLE
#         prospects_report = soup.select('#portal-prospects-report > table')        
#         for div in prospects_report:
#             fv = div.findAll('td')[5].text
            
        # Dashoard Stats
        dashboard_stats = soup.select('#dashboard > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in dashboard_stats:
            
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    war = stats.findAll('td', {'data-stat': 'WAR'})[0].text
                    total_war += float(war)
        
        # Pitch Info Pitch Velocity Stats
        pipv_stats = soup.select('#pitch-info-pitch-velocity > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in pipv_stats:
            
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    fb_velo = stats.findAll('td', {'data-stat': 'vFA'})[0].text
        
        # Standard Stats
        standard_stats = soup.select('#standard > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in standard_stats:

            # For each row of data that has Milb stats
            for stats in div.findAll('tr', {'class': 'row-minors is-selected__invalid'}):

                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    w = stats.findAll('td', {'data-stat': 'W'})[0].text
                    sv = stats.findAll('td', {'data-stat': 'SV'})[0].text
                    hld = stats.findAll('td', {'data-stat': 'HLD'})[0].text
                    ip = stats.findAll('td', {'data-stat': 'IP'})[0].text
                    tbf = stats.findAll('td', {'data-stat': 'TBF'})[0].text
                    h = stats.findAll('td', {'data-stat': 'H'})[0].text                    
                    er = stats.findAll('td', {'data-stat': 'ER'})[0].text                    
                    hra = stats.findAll('td', {'data-stat': 'HR'})[0].text
                    bb = stats.findAll('td', {'data-stat': 'BB'})[0].text
                    ibb = stats.findAll('td', {'data-stat': 'IBB'})[0].text                    
                    hbp = stats.findAll('td', {'data-stat': 'HBP'})[0].text                    
                    so = stats.findAll('td', {'data-stat': 'SO'})[0].text
                    f_ip = str(ip).split('.')[0]
                    full_ip += int(f_ip)
                    p_ip = str(ip).split('.')[1]
                    partial_ip += int(p_ip)
                    total_w += int(w)
                    total_sv += int(sv)
                    total_hld += int(hld)
                    total_bf += int(tbf)
                    total_ha += int(h)                    
                    total_er += int(er)
                    total_hra += int(hra)                    
                    total_bba += int(bb)
                    total_ibba += int(ibb)
                    total_hbpa += int(hbp)
                    total_so += int(so)                    
                    
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    w = stats.findAll('td', {'data-stat': 'W'})[0].text                    
                    sv = stats.findAll('td', {'data-stat': 'SV'})[0].text
                    hld = stats.findAll('td', {'data-stat': 'HLD'})[0].text                    
                    ip = stats.findAll('td', {'data-stat': 'IP'})[0].text
                    tbf = stats.findAll('td', {'data-stat': 'TBF'})[0].text
                    h = stats.findAll('td', {'data-stat': 'H'})[0].text                                        
                    er = stats.findAll('td', {'data-stat': 'ER'})[0].text
                    hra = stats.findAll('td', {'data-stat': 'HR'})[0].text    
                    bb = stats.findAll('td', {'data-stat': 'BB'})[0].text
                    ibb = stats.findAll('td', {'data-stat': 'IBB'})[0].text                        
                    hbp = stats.findAll('td', {'data-stat': 'HBP'})[0].text                    
                    so = stats.findAll('td', {'data-stat': 'SO'})[0].text                
                    f_ip = str(ip).split('.')[0]
                    full_ip += int(f_ip)
                    p_ip = str(ip).split('.')[1]
                    partial_ip += int(p_ip)
                    total_w += int(w)
                    total_sv += int(sv)
                    total_hld += int(hld)                    
                    total_bf += int(tbf)
                    total_ha += int(h)
                    total_er += int(er)                    
                    total_hra += int(hra)
                    total_bba += int(bb)
                    total_ibba += int(ibb)                    
                    total_hbpa += int(hbp)                    
                    total_so += int(so)
                    
        # Batted Ball Stats
        battedball_stats = soup.select('#batted-ball > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in battedball_stats:
            
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:
                    
                    # Stats
                    siera = stats.findAll('td', {'data-stat': 'SIERA'})[0].text
                    
        # More Batted Ball Stats
        morebattedball_stats = soup.select('#more-batted-ball > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in morebattedball_stats:
            
            # For each row of data that has MiLB stats
            for stats in div.findAll('tr', {'class': 'row-minors is-selected__invalid'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:
                    
                    # Stats
                    gb = stats.findAll('td', {'data-stat': 'GB'})[0].text
                    fb = stats.findAll('td', {'data-stat': 'FB'})[0].text                    
                    ld = stats.findAll('td', {'data-stat': 'LD'})[0].text                  
                    total_gb += int(gb)
                    total_fb += int(fb)                    
                    total_ld += int(ld)
            
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:
                    
                    # Stats
                    gb = stats.findAll('td', {'data-stat': 'GB'})[0].text
                    fb = stats.findAll('td', {'data-stat': 'FB'})[0].text                    
                    ld = stats.findAll('td', {'data-stat': 'LD'})[0].text
                    total_gb += int(gb)
                    total_fb += int(fb)                    
                    total_ld += int(ld)                    

        #Calc Stats
        total_ip = round(float(full_ip) + float(partial_ip/3), 2) 
        
        try:
            era = round((total_er / total_ip) * 9, 2)
        except:
            era = 0
        try:
            whip = round((total_bba + total_ha) / total_ip, 2)
        except:
            whip = 0
        try:
            so_percent = round((total_so / total_bf * 100), 1)
        except:
            so_percent = 0
        try:
            bb_percent = round((total_bba / total_bf * 100), 1)
        except:
            bb_percent = 0
        try:
            fip = round((((13 * total_hra) + (3 * (total_bba + total_hbpa)) - (2 * total_so)) / total_ip) + cFIP, 2)
        except:
            fip = 0
        try:
            xfip = round((((13 * (total_fb * lgHR_FB_percent)) + (3 * (total_bba + total_hbpa)) - (2 * total_so)) / total_ip) + cFIP, 2)
        except:
            xfip = 0
        try:
            k9 = round((total_so / total_ip) * 9, 2)
        except:
            k9 = 0
        try:
            walk9 = round((total_bba / total_ip) * 9, 2)
        except:
            walk9 = 0
        try:
            kbb = round(total_so / total_bba, 2)
        except:
            kbb = 0
                   
        
        #Create Player Object
        player_table.append({
            "Name": name[loop_counter], 
            #"Team": team,
            "Fangraphs Link": url,
            "Owner": owner[loop_counter],
            "Fangraphs Name": fg_name,           
            "DOB": player_dob,
            "Position": position,
            "Level": level,
            "WAR": total_war,
            "IP": total_ip,
            "W": total_w,
            "SO": total_so,
            "SV": total_sv,
            "HLD": total_hld,
            "ERA": era,
            "WHIP": whip,
            "BB": total_bba,
            "SO%": f'{so_percent}%',
            "K/9": k9,
            "WALK%": f'{bb_percent}%',
            "WALK/9": walk9,
            "K/BB": kbb,
            "FIP": fip,
            "xFIP": xfip,
            "SIERA": siera,
            "FB Velo": fb_velo
        })

    
    
    #################################
    ######### H I T T E R S #########
    #################################
    
    else:
    #if 'P' not in position:
                      
        # Dashboard Stats
        dashboard_stats = soup.select('#dashboard > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in dashboard_stats:
                
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    war = stats.findAll('td', {'data-stat': 'WAR'})[0].text
                    total_war += float(war)   
        
        # Standard Stats
        standard_stats = soup.select('#standard > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in standard_stats:

            # For each row of data that has Milb stats
            for stats in div.findAll('tr', {'class': 'row-minors is-selected__invalid'}):

                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    ab = stats.findAll('td', {'data-stat': 'AB'})[0].text  
                    pa = stats.findAll('td', {'data-stat': 'PA'})[0].text                      
                    singles = stats.findAll('td', {'data-stat': '1B'})[0].text                    
                    doubles = stats.findAll('td', {'data-stat': '2B'})[0].text                     
                    triples = stats.findAll('td', {'data-stat': '3B'})[0].text                     
                    hrs = stats.findAll('td', {'data-stat': 'HR'})[0].text                     
                    r = stats.findAll('td', {'data-stat': 'R'})[0].text  
                    rbi = stats.findAll('td', {'data-stat': 'RBI'})[0].text 
                    bb = stats.findAll('td', {'data-stat': 'BB'})[0].text                      
                    ibb = stats.findAll('td', {'data-stat': 'IBB'})[0].text
                    so = stats.findAll('td', {'data-stat': 'SO'})[0].text                      
                    hbp = stats.findAll('td', {'data-stat': 'HBP'})[0].text
                    sf = stats.findAll('td', {'data-stat': 'SF'})[0].text                    
                    sb = stats.findAll('td', {'data-stat': 'SB'})[0].text  
                    total_h_ab += int(ab)
                    total_h_pa += int(pa)                      
                    total_singles += int(singles)
                    total_doubles += int(doubles)
                    total_triples += int(triples)
                    total_hrs += int(hrs)  
                    total_h_r += int(r)
                    total_rbi += int(rbi)
                    total_h_bb += int(bb)
                    total_h_ibb += int(ibb)
                    total_h_so += int(so)                    
                    total_hbp += int(hbp)
                    total_sf += int(sf)
                    total_sb += int(sb)
                    
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    ab = stats.findAll('td', {'data-stat': 'AB'})[0].text                        
                    pa = stats.findAll('td', {'data-stat': 'PA'})[0].text                    
                    singles = stats.findAll('td', {'data-stat': '1B'})[0].text                    
                    doubles = stats.findAll('td', {'data-stat': '2B'})[0].text                     
                    triples = stats.findAll('td', {'data-stat': '3B'})[0].text                     
                    hrs = stats.findAll('td', {'data-stat': 'HR'})[0].text    
                    r = stats.findAll('td', {'data-stat': 'R'})[0].text  
                    rbi = stats.findAll('td', {'data-stat': 'RBI'})[0].text 
                    bb = stats.findAll('td', {'data-stat': 'BB'})[0].text                      
                    ibb = stats.findAll('td', {'data-stat': 'IBB'})[0].text                       
                    so = stats.findAll('td', {'data-stat': 'SO'})[0].text                     
                    hbp = stats.findAll('td', {'data-stat': 'HBP'})[0].text
                    sf = stats.findAll('td', {'data-stat': 'SF'})[0].text     
                    sb = stats.findAll('td', {'data-stat': 'SB'})[0].text  
                    total_h_ab += int(ab)
                    total_h_pa += int(pa)                    
                    total_singles += int(singles)
                    total_doubles += int(doubles)
                    total_triples += int(triples)
                    total_hrs += int(hrs) 
                    total_h_r += int(r)
                    total_rbi += int(rbi)
                    total_h_bb += int(bb)
                    total_h_ibb += int(ibb)
                    total_h_so += int(so)
                    total_hbp += int(hbp)
                    total_sf += int(sf)
                    total_sb += int(sb)
        
        #Advanced Stats
        advanced_stats = soup.select('#advanced > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in advanced_stats:

            # For each row of data that has Milb stats
            for stats in div.findAll('tr', {'class': 'row-minors is-selected__invalid'}):

                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    wrc = stats.findAll('td', {'data-stat': 'wRC'})[0].text
                    total_wrc += int(wrc)
                    
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    wrc = stats.findAll('td', {'data-stat': 'wRC'})[0].text
                    total_wrc += int(wrc)
                    
        # Statcast Stats
        statcast_stats = soup.select('#statcast > div:nth-child(3) > div > div > div > div.table-fixed > table')
        for div in statcast_stats:
                    
            # For each row of data that has MLB stats
            for stats in div.findAll('tr', {'class': 'row-mlb-season'}):
                
                # Create/clear yearly stats
                stats_year = stats.findAll('td', {'data-stat': 'Season'})[0].text

                # Get totals for last year
                if stats_year == current_year:

                    # Stats
                    events = stats.findAll('td', {'data-stat': 'Events'})[0].text
                    ev = stats.findAll('td', {'data-stat': 'EV'})[0].text
                    la = stats.findAll('td', {'data-stat': 'LA'})[0].text
                    barrels = stats.findAll('td', {'data-stat': 'Barrels'})[0].text
                    hardhit = stats.findAll('td', {'data-stat': 'HardHit'})[0].text
                    total_events += int(events)                    
                    total_barrels += int(barrels)                    
                    total_hardhit += int(hardhit) 
                    hardhit_percent = round(total_hardhit / total_events * 100, 1)
                    barrel_percent = round(total_barrels / total_events * 100, 1)
                    
        # Calc Stats
        total_h_hits = total_singles + total_doubles + total_triples + total_hrs
        
        try:
            woba = (wbb * (total_h_bb - total_h_ibb) + whbp * total_hbp + w1b * total_singles + w2b * total_doubles + w3b * total_triples + whr * total_hrs) / (total_h_ab + total_h_bb - total_h_ibb + total_sf + total_hbp)
        except:
            woba = 0
        try:
            iso = round((total_doubles + 2 * total_triples + 3 * total_hrs) / total_h_ab, 3)
        except:
            iso = 0
        try:
            k_percent = round(total_h_so / total_h_pa * 100, 1)
        except:
            k_percent =0
        try:
            bb_percent = round(total_h_bb / total_h_pa * 100, 1)
        except:
            bb_percent = 0
        try:
            ops = round(((total_h_hits + total_h_bb + total_hbp) / (total_h_ab + total_h_bb + total_hbp + total_sf)) + ((total_singles + 2 * total_doubles + 3 * total_triples + 4 * total_hrs) / total_h_ab), 3)
        except:
            ops = 0
        try:
            avg = round(total_h_hits / total_h_ab, 3)
        except:
            avg = 0
#         try:
#             wrc = (((woba - league_woba) / woba_scale) + (r_pa)) * total_h_pa
#         except:
#             wrc = 0
        try:
            wrc400 = (total_wrc / total_h_pa) * 400
        except:
            wrc400 = 0
        
        #Create Player Object        
        player_table.append({
        "Name": name[loop_counter],     
        #"Team": team,
        "Fangraphs Link": url,
        "Owner": owner[loop_counter],
        "Fangraphs Name": fg_name,   
        "DOB": player_dob,
        "Position": position,
        "Level": level,
        "WAR": total_war,
        "AB": total_h_ab,
        "R": total_h_r,
        "HR": total_hrs,
        "RBI": total_rbi,
        "SB": total_sb,
        "AVG": avg,
        "OPS": ops,
        "ISO": iso,
        "wOBA": round(woba, 3),
        "K%": f'{k_percent}%',
        "BB%": f'{bb_percent}%',
        "EV": ev,
        "LA": la,
        "HardHit%": f'{hardhit_percent}%',
        "Barrel%": f'{barrel_percent}%',
        "wRC": round(total_wrc, 0),
        "wRC400": round(wrc400, 0)
        })
    
    print(name[loop_counter], fg_name)
    
    loop_counter += 1
    
```

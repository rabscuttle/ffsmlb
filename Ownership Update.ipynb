```python
# Import
import pandas as pd
```

```python
# Variables
file_name = "FFS Spreadsheet.xlsx"
ffs_teams_list = ['ARI','ATL','BAL','BOS','CHC','CHW','CIN','CLE','COL','DET','HOU','KC','LAA','LAD','MIA','MIL','MIN','NYM','NYY','OAK','PHI','PIT','SD','SEA','SF','STL','TB','TEX','TOR','WAS']
current_year = "2023"
```

```python
# Read csv
csv_player_list = pd.read_csv('playerownership.csv', encoding = "ISO-8859-1")
csv_names = csv_player_list['name'].tolist()
csv_alt_names = csv_player_list['alt_name'].tolist()
csv_owners = csv_player_list['owner'].tolist()
print(csv_player_list)
```

```python
# Import FFS MLB Sheet
# Read Player Names
name_cols = "G"
change_list = []
sheet_players = []
sheet_owners = []

# Loop through each team to create the sheet_players list
for team in ffs_teams_list:
    # Variables
    sheet = pd.read_excel(file_name, sheet_name=team, usecols=name_cols, skiprows=2)
    sheet_names = sheet['Player']

    # Remove empty rows
    nan_value = float("NaN")
    sheet_names.replace("", nan_value, inplace=True)
    sheet_names.dropna(inplace=True)

    # Write progress to console
    #print(f"Processing data for {team}")

    # Loop through each name in the selected team sheet
    for sheet_name in sheet_names:

        # Remove FFS lingo from name (formatting)
        sheet_name = sheet_name.replace('RV', '')
        sheet_name = sheet_name.replace('R1', '')
        sheet_name = sheet_name.replace(team, '')
        sheet_name = sheet_name.strip()
        sheet_players.append(sheet_name)
        sheet_owners.append(team)

# Loop through each player in csv
loop_counter = 0
for name in csv_names:
    
    alt_name = csv_alt_names[loop_counter]
    owner = csv_owners[loop_counter]
    
    # Loop through sheet_players
    sheet_loop_counter = 0
    for sheet_name in sheet_players:
            
        # If player name matches
        sheet_owner = sheet_owners[sheet_loop_counter]        
        if sheet_name == name or sheet_name == alt_name:
            if sheet_owner == owner:
                change_list.append({
                    "Name": name, 
                    "Alt Name": alt_name,
                    "Sheet Name": sheet_name,
                    "Matched": "Matched",
                    "New Owner": owner
                })      
                break
            
            # But the owner isn't the same
            if sheet_owner != owner:
                print(f'Moving {name} from {owner} to {sheet_owner}.')
                change_list.append({
                    "Name": name, 
                    "Alt Name": alt_name,
                    "Sheet Name": sheet_name,
                    "Matched": "New Owner",
                    "New Owner": sheet_owner,
                    "Old Owner": owner
                })                         
                break
        
        # If name doesn't match and the loop has finished
        if sheet_name != name and sheet_name != alt_name:
            if sheet_loop_counter == len(sheet_players) - 1:
                print(f'{name} not matched.')
                change_list.append({
                    "Name": name, 
                    "Alt Name": alt_name,
                    "Old Owner": owner,
                    "Matched": "Not found in sheet"
                })      
                break    
                
        sheet_loop_counter += 1
    loop_counter += 1
                
```

```python
ownership_df = pd.DataFrame(change_list)
ownership_df.to_csv(f"players/ownership.csv", index = False, header=True)
ownership_df
```

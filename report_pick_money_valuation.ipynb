```python
import pandas as pd
import numpy as np
```

```python
# Import csv
csv = pd.read_csv('draft/pickmoneytrades.csv') 
```

```python
# One Pick for Money
one_pick_for_money = csv[csv.Team1Picks1.notnull() & csv.Team1Picks2.isnull() & csv.Team2Picks1.isnull()]
other_way = csv[csv.Team2Picks1.notnull() & csv.Team2Picks2.isnull() & csv.Team1Picks1.isnull()]
one_pick_for_money.append(other_way)
single_pick_value = one_pick_for_money.groupby(by="Team1Picks1").mean().sort_values('Team2Money', ascending=False)
single_pick_value = single_pick_value.drop(columns=['Team1Picks5', 'Team1Picks6', 'Team1Money', 'Team2Picks5', 'Team2Picks6'])
single_pick_value['pick_number'] = single_pick_value.index.astype(str)
single_pick_value

```

```python
# Yearly tables
# This year
picks_array = np.arange(1,200)
pick_val_chart_df = pd.DataFrame(data=picks_array, columns=['pick_number'])
pick_val_chart_df['pick_number'] = pick_val_chart_df['pick_number'].astype(str)
pick_val_chart_df = pd.merge(pick_val_chart_df, single_pick_value, on='pick_number', how='outer')
pick_val_chart_df = pick_val_chart_df.rename(columns={"Team2Money": "money"})
pick_val_chart_df
```
